```python

from llama_index.prompts.prompts import SimpleInputPrompt


system_prompt = "You are a Q&A assistant. Your goal is to answer questions as accurately as possible based on the instructions and context provided."


query_wrapper_prompt = SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
import torch

llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="meta-llama/Llama-2-7b-chat-hf",
    model_name="meta-llama/Llama-2-7b-chat-hf",
    device_map="auto",
    model_kwargs={"torch_dtype": torch.float16 , "load_in_4bit":True}
)
```

```python
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from llama_index import LangchainEmbedding, ServiceContext

embed_model = LangchainEmbedding(
HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
)
```

```python
!nvidia-smi
```

```python
service_context = ServiceContext.from_defaults(
    chunk_size=4096,
    llm=llm,
    embed_model=embed_model
)
```

**Here, we are using 10,000 movie data from kaggle, Please find the DataSet link ** : https://www.kaggle.com/datasets/willianoliveiragibin/10000-data-about-movies-1915-2023/


```python
import pandas as pd
df = pd.read_csv('data.csv')
display(df.head(1), df.shape)
```

```python
df = df.drop("Unnamed: 0", axis=1)
```

**Preparing Data to Insert (Preprocessing)**

```python
def preprocess(row):

  docs = []

  movie=row["Movie Name"]
  year=row["Year of Release"]
  rating=float(row["Movie Rating"])
  genre = row['Genre']
  director = row['Director']
  starts = row['Stars']
  description=str(row["Description"])


  row['page_content'] = f"""
  movie: {movie}
  year: {year}
  rating: {rating}
  genre: {genre}
  director: {director}
  starts: {starts}
  description: {description}
  """
  row['metadata']={"movie": movie,
            "year": year,
            "rating": rating,
            "genre": genre,
            "director": director,
            "starts": starts,
            "description":description
            }

  return row

df = df.apply(preprocess, axis=1)
df.head(2)
```

```python
from llama_index.schema import Document
```

```python
documents = []
```

# Adding page content and meta data in list 'document'

```python
def documents_parser(row):

  data = Document(text=row['page_content'],
        metadata=row['metadata'])
  documents.append(data)

df.apply(documents_parser, axis=1)
```

```python
len(documents)
```

**Indexing in vectorStoreIndex using llama index**

```python
index = VectorStoreIndex.from_documents(documents, service_context=service_context)
```

```python
import locale
locale.getpreferredencoding = lambda: "UTF-8"
```

# Q&A with Indexed Data

```python
query_engine = index.as_query_engine(local=True)
query_string = "Can you please let me know the year of release for the movie name 'The Shawshank Redemption'?"
response = query_engine.query(query_string)
response.response
```
